# 사용자 위치와 지하철 api를 이용한 지하철 승하차 알리미 앱

# 주제 선정 이유
  

지하철이 없는 곳에서 올라와 기숙사 생활을 하고있다. 그러다 보니 안 타던 지하철을 타면서 느낀 불편한 점이 있다. 한때는 중고거래를 하러 영통역에서 수인분당선을 타고 신갈로 가야했다. 난 이어폰을 쓰고 있었고 안내방송을 제대로 듣지 못했다. 열차 창문이 스크린 도어 광고로 인해 가려져서 역 명도 안 보이고 차내에 있는 스크린에도 역 명을 한번 보여주고 만다. 그래서 난 어렴풋이 들은 '갈'이라는 소리를 듣고 내렸다. 하지만 그곳은 상갈이였고 결국 시간을 맞추기 위해 택시를 타야 했다. 요즘 지하철에는 이어폰은 물론이고 스마트폰으로 동영상을 보거나 소음 차단 이어폰을 쓰는 경우도 많다. 마냥 안내 방송을 못들은 사람만을 탓하고 있을 수는 없다. 그러다 난 문득 코레일 톡 앱에 있는 열차의 실시간 운행 정보를 보게 되었고 이를 탑승하는 사람의 시점으로 잘 적용해서 지하철 앱으로 만들어 보고 싶다고 생각했다. 

# 가설 정의


도시철도 초심자 뿐만 아니라 지하철에서 영상을 시청하거나 소음 차단 이어폰을 사용하는 사람들에게 도움이 될 것이라고 생각한다. 또한 이를 버스에 적용하여 지하철과 연계를 하면 더 실용적인 앱이 될 것이다. 오히려 실시간 위치를 제공하는 버스에 대해 앱을 만드는 것이 더 수월할 수 있다. 지하철 역별로 도착정보만 나와있는 경우에는 그 도착정보를 통하여 자체적으로 열차의 위치를 구하는 프로그램을 만들 필요성이 있다.

# 개발 계획

개발은 두가지 언어로 할 계획이다. 먼저 앱 개발을 해야하기 때문에 코틀린이라는 언어를 안드로이드 스튜디오라는 IDE로 쓸 계획이다. 그리고 공공 api를 사용하여 앱에서 쓸 데이터를 가공하는 작업을 파이썬으로 할 예정이다. 그러기 위해서는 파이썬으로 공공 api를 사용하는 방법을 인터넷을 통해 공부하고 책으로 코틀린을 공부해야 한다. 안드로이드 개발 언어로 코틀린을 사용하는 이유는 뒤늦게 주제를 바꾸어서 기존 동아리 대항전에서 준비하는 프로젝트와 배울 언어를 똑같이 맞추는 방식으로 시간을 줄이기 위함이다. 그리고 파이썬으로 가공한 데이터를 구글 파이어베이스를 이용하여 연동할 예정이다. 그리고 파이썬으로 가공된 데이터를 파이어 베이스에 연동시킬 것이다. 구체적인 계획 수립이나 수정은 개발 도중에 할 예정이다. 

서울시 공공데이터 포털에서 두가지 api를 사용할 예정이다. 한 개는 역별 실시간 지하철 도착정보이고 한 개는 실시간 지하철 위치 정보이다.
